```python
import re  
import json  

# 原始文本数据  
text = steps
# 创建一个空字典用于保存步骤  
steps_d = {}  

# 使用正则表达式解析每一行  
lines = text.split('\n')  # 按行分割文本  
for line in lines:  
    match = re.match(r"^(步骤\d+)：（(.*)）$", line)  
    if match:  
        step_key = match.group(1)  
        step_value = match.group(2).split('，')  # 将内容按逗号分割解析为列表  
        steps_d[step_key] = step_value

# 将字典转换为 JSON 格式字符串  
json_data = json.dumps(steps_d, ensure_ascii=False, indent=4)  

# 输出 JSON 数据  
print(json_data)  

# 可将 JSON 写入文件  
with open("steps.json", "w", encoding="utf-8") as f:  
    f.write(json_data)
```

```python
class Experimental:  

    def __init__(self, llm):  
        self.llm = llm  
        self.experimental_steps_chain()  

    def __repr__(self):  
        return f"Experimental({self.llm})"  

    def __call__(self, experimental):  
        # 综合运行链条  
        self.combine_chain()  

        # 传递输入并运行  
        # `apply` 是用于多输入和多输出的支持  
        result = self.sequential_chain.batch([{"experimental": experimental}])[0]  
        
        # 获取多个结果变量  
        print(f"实验步骤: {result['experimental_steps']}")  
        print(f"生成的网络图: {result['steps']}")  
        print(f"校验结果: {result['check_result']}")  
        return result

    # 定义所有链条  
    def experimental_steps_chain(self):  
        # 第一个链: 转化实验描述到实验步骤  
        experimental_template = """  
        根据实验信息：  
        {experimental}  
        输出实验所需的步骤，格式如下：  
        step1: ...;  
        step2: ...;  
        step3: ...;  
        """  
        experimental_prompt = PromptTemplate(  
            input_variables=["experimental"],   
            template=experimental_template  
        )  
        self.experimental_chain = LLMChain(  
            llm=self.llm,   
            prompt=experimental_prompt,   
            output_key="experimental_steps"  
        )  

        # 第二个链: 转化实验步骤到图结构  
        steps_template = """  
        实验步骤如下：  
        {experimental_steps}  
        请将其转化成网络图，输出格式如下：  
        {{  
            "nodes": [  
                {{"id": "节点1", "label": "节点标签"}}  
            ],  
            "edges": [  
                {{"source": "节点1", "target": "节点2", "relationship": "关系"}}  
            ]  
        }}  
        如：火柴1点燃酒精灯
        {{  
            "nodes": [  
                {{"id": "1", "label": "火柴1"}}  
                {{"id": "3", "label": "酒精灯"}}  

            ],  
            "edges": [  
                {{"source": "1", "target": "3", "relationship": "点燃"}}  
            ]  
        }}  
        可以看到，节点都是实验具体器材，边包含操作信息
        """  
        steps_prompt = PromptTemplate(  
            input_variables=["experimental_steps"],  
            template=steps_template  
        )  
        self.steps_chain = LLMChain(  
            llm=self.llm,   
            prompt=steps_prompt,   
            output_key="steps"  
        )  

        # 第三个链: 校验准确性  
        check_template = """  
        实验步骤：  
        {experimental_steps}  
        生成的图：  
        {steps}  
        请判断图和步骤描述是否对应，返回yes/no：  
        """  
        check_prompt = PromptTemplate(  
            input_variables=["experimental_steps", "steps"],  
            template=check_template  
        )  
        self.check_chain = LLMChain(  
            llm=self.llm,   
            prompt=check_prompt,   
            output_key="check_result"  
        )  

    def combine_chain(self):  
        # 链条组合  
        chains = [self.experimental_chain, self.steps_chain, self.check_chain]  
        self.sequential_chain = SequentialChain(  
            chains=chains,   
            input_variables=["experimental"],   
            output_variables=["experimental_steps", "steps", "check_result"]  
        )
```
